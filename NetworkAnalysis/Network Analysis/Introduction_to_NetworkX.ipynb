# Introduction to NetworkX

[NetworkX](https://networkx.org) is a Python package used for creating, loading in, and studying graphs and networks. In this notebook we will explore the basic functionality of NetworkX v.2.6.0, including:

* Creating and modifying simple undirected graphs
* Accessing simple graph attributes
* Drawing and formating graphs



---

```python
# Import the necessary python packages
import networkx as nx
import matplotlib.pyplot as plt
```

---

The simpliest way to understand the NetworkX graph class is to instantiate an empty graph object with no nodes and no edges. 

---

```python
# Initialize a graph
G = nx.Graph()
print(f"The type of G is: {type(G)} \n")
```

---

It is important to remember that Python objects have attributes and methods. A full list of these properties can be found at [NetworkX graph documentation](https://networkx.org/documentation/stable/reference/classes/graph.html#networkx.Graph). The following cell illustrates the basic properties of the NetworkX graph class. 

---

```python
# What are the vertices?
print(f"The nodes of G are: V(G) = {G.nodes} \n")

# What are the edges?
print(f"The edges of G are: E(G) = {G.edges} \n")
```

---

Let us now add nodes to ``G`` using the ``G.add_node()`` method. 

---

```python
G.add_node("a")
G.add_node("b")
G.add_node("c")
G.add_node("d")

# What are the vertices?
print(f"The nodes of G are: V(G) = {G.nodes} \n")

# What are the edges?
print(f"The edges of G are: E(G) = {G.edges} \n")
```

---

Let us now add edges to ``G`` using the ``G.add_edge()`` method. 

---

```python
G.add_edge("a", "b")
G.add_edge("a", "c")
G.add_edge("b", "c")
G.add_edge("c", "d")

# What are the vertices?
print(f"The nodes of G are: V(G) = {G.nodes} \n")

# What are the edges?
print(f"The edges of G are: E(G) = {G.edges} \n")
```

---

We can view the number of nodes and number of edges in ``G`` with NetworkX by calling the ``nx.number_of_nodes()`` and ``nx.number_of_edges()`` functions, respectively. 

---

```python
# What are the vertices?
print(f"The number of nodes in G is: {nx.number_of_nodes(G)} \n")

# What are the edges?
print(f"The number of edges in G is: {nx.number_of_edges(G)} \n")
```

---

NetworkX provides several functions for drawing graphs. The most basic such function is the ``nx.draw()`` function and is illustrated with the following code cell. 

---

```python
# A basic drawing function for viewing a graph
nx.draw(G)
```

---

Viewing graphs can be an important aspect of network analysis and ``nx.draw()`` gives us basic functionality to do so. However, this type of drawing is very limited when viewing large graphs. For example, consider this function applied to a [Barabasi Albert_graph](https://networkx.org/documentation/stable/reference/generated/networkx.generators.random_graphs.barabasi_albert_graph.html#networkx.generators.random_graphs.barabasi_albert_graph) with 150 nodes. 

---

```python
# Instanstiate a Barabasi Albert graph on 150 nodes 
H = nx.barabasi_albert_graph(350, m=2, seed=1)

# Try drawing H
nx.draw(H)
```

---

The NetworkX drawing functionality is built on the well known [matplotlib](https://matplotlib.org) Python package. Knowing this, we can take our time and slowly construct a better drawing of ``H`` above. Through my own experiences, I found that this process can best be constructed by following the steps below:


1. Set your figure size using ``plt.figure(figsize=(12, 10))`` function from ``matplotlib.pytplot``.


2. Set ``pos = nx.draw_kamada_kawai(graph)``


3. Set how your nodes should appear with a Python dictionary. I typically call this dictionary ``node_options``. This will be passed into the NetworkX function ``nx.draw_networkx_nodes``, and will only draw the nodes of the graph  according to what you have specified in ``node_options``. 


4. Set how your edges should appear with a Python dictionary. I typically call this dictionary ``edge_options``. This will be passed into the NetworkX function ``nx.draw_networkx_edges``, and will only draw the edges of the graph according to what you have specified in ``edge_options``. 


5. Call the ``plt.show()`` function from ``matplotlib.pytplot``.



The following code cell illustrates what is meant by these steps. I encourage you to remove and change the code in order to fully understand how the code is working! 


---

```python
# Visualizing graphs is not the easiest!


# Set your figure size
plt.figure(figsize=(10, 8))

# Set your pos
pos = nx.kamada_kawai_layout(H)

# Set your node options
node_options = {"node_color": "green", "node_size": 80}

# Set your edge options
edge_options = {"width": 1.0, "alpha": .25, "edge_color": "purple"}

# Draw nodes
nx.draw_networkx_nodes(H, pos, **node_options)

# Draw edges
nx.draw_networkx_edges(H, pos, **edge_options)


plt.show()
```

---

Let us now try drawing the graph ``G`` we created at the beginning of this notebook!

---

```python
# Set your figure size
plt.figure(figsize=(10, 8))

# Set your pos
pos = nx.kamada_kawai_layout(G)

# Set your node options
node_options = {"node_color": "black", "node_size": 30}

# Set your edge options
edge_options = {"width": .50, "alpha": .5, "edge_color": "black"}

# Draw nodes
nx.draw_networkx_nodes(G, pos, **node_options)

# Draw edges
nx.draw_networkx_edges(G, pos, **edge_options)
plt.show()
```

---

The drawing above seems satisfactory. However, the nodes of our graph have labels, namely ``"a", "b", "c", `` and ``"d"``. We can draw these labels by creating a Python dictionary describing the features of our labels, i.e., text color, font size, and allignment. I typically call this dictionary ``node_label_options``. These options can then be passed into ``nx.draw_network_labels`` as the following code cell illustrates. 


---

```python
# Set your figure size
plt.figure(figsize=(10, 8))

# Set your pos
pos = nx.kamada_kawai_layout(G)

# Set your node options
node_options = {"node_color": "black", "node_size": 30}

# Set your edge options
edge_options = {"width": .50, "alpha": .5, "edge_color": "black"}

# Set your node label options
node_label_options = {"font_size": 15,
                      "font_color": "blue",
                      "verticalalignment": "bottom",
                      "horizontalalignment": "left"}

# Draw nodes
nx.draw_networkx_nodes(G, pos, **node_options)

# Draw edges
nx.draw_networkx_edges(G, pos, **edge_options)

# Draw your labels
nx.draw_networkx_labels(G, pos, **node_label_options)

plt.show()
```

```python
G = nx.Graph()

edges = [("Ben", "Randy"),
         ("Ben", "Andrew"),
         ("Randy", "Andrew"),
         ("Randy", "Morgan"),
         ("Randy", "David"),
         ("Andrew", "Daina"),
         ("Daina", "Morgan"),
         ("Andrew", "David"),
         ("Andrew", "Morgan"),
         ("Andrew", "Jeremy")]

G.add_edges_from(edges)

print(f"The nodes of G are: {G.nodes} \n")
print(f"The edges of G are: {G.edges} \n")
```

```python
# Set your figure size
plt.figure(figsize=(10, 8))

# Set your pos
pos = nx.kamada_kawai_layout(G)

# Set your node options
node_options = {"node_color": "red", "node_size": 30}

# Set your edge options
edge_options = {"width": 1.50, "alpha": .5, "edge_color": "green"}

# Set your node label options
node_label_options = {"font_size": 15,
                      "font_color": "blue",
                      "verticalalignment": "bottom",
                      "horizontalalignment": "left"}

# Draw nodes
nx.draw_networkx_nodes(G, pos, **node_options)

# Draw edges
nx.draw_networkx_edges(G, pos, **edge_options)

# Draw your labels
nx.draw_networkx_labels(G, pos, **node_label_options)

plt.show()
```
