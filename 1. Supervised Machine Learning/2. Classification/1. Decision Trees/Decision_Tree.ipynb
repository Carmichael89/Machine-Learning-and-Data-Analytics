In this module, we will utilize the [DecisionTreeClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html) module from sci-kit learn. This will allow us to rapidly develop a model and use the built in model metrics to develop and compare our model quickly and easily.

This project will follow the 4 phase process of a typical machine learning project.

Phase 2: Train the model

Phase 3: Test the model

# Phase 1 - Data Cleaning/Analysis


```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
from sklearn.metrics import ConfusionMatrixDisplay

data_df = pd.read_csv('diabetes_data_upload.csv')

data_df.head()
```

Above is a snippet of our un-modified data as a pandas dataframe. This data is not in a useful format and needs to be changed to be used in a machine learning model. We will start by encoding each feature with a 0 or 1 instead of a string to represent postive/negative, male/female, etc. We will also standardize the age value to provide a more consistent value range with the rest of the features. In this module, we will use scikit-learn's standardscalar function to cleanly and easily standardize the values. Finally, we need to append a column of ones to act as the bias of our model.

```python
data_df = data_df.replace(['Male', 'Yes', 'Positive'], 1)
data_df = data_df.replace(['Female', 'No', 'Negative'], 0)
data_df.head()
```

```python
scaler = StandardScaler()
data_df[['Age']] = scaler.fit_transform(data_df[['Age']])
data_df.head()
```

Above is the results of our cleaning. The data is now useful, the bias is now a part of our features, and the age feature is normalized. 

---
Next, we need to shuffle and split the data, train and test our model, then view the metrics. Using sci-kit learn, this process becomes incredible simple. The [scikit-learn website](https://scikit-learn.org/stable/) provides a model, a data shuffle and split method, and convenient methods to display the confusion matrix and f score. 

```python
X_train, X_test, y_train, y_test = train_test_split(data_df.iloc[:,:-1], data_df.iloc[:,-1], test_size = int(data_df.shape[0]*0.1), train_size = int(data_df.shape[0]*0.9), random_state=3)
print(X_train.shape)
print(X_test.shape)
```

# Phase 2: Train the model
In this phase, we need to prepare the data for the training instance and train the model. Next, we will create a decision tree model using sci-kit learn's DecisionTreeClassifier module and fit the data to the model. Here, we create a model using the default values which do not set restrictions on the size of the model. This dataset is relatively small and we are interested in the best performing model, therefore we do not want to restrict the size or complexity of the model, except for the case of overfitting, which can be evaluated against the test set. Using sci-kit Learn is very intuitive and our model can be created and trained with two lines of code as seen below.

```python
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
```

# Phase 3: Test the Model
Our decision tree classifier has been created and trained against our training data. Now, we will evaluate its performance against the reserved set of testing data using the F score and confusion matrix. A benifit of using a professional library to create machine learning algorithms is evaluation metrics are typically included and only require a function call and data as seen below. Here, we will also provide a visual display of our decision tree using the "plot_tree" method from sci-kit learn. This will allow us to see how complex and deep our model is.

```python
y_hat = decision_tree.predict(X_test)
print("F1 score: ", np.round(f1_score(y_test, y_hat),2))

plt.figure(figsize = (50,50))

a = tree.plot_tree(decision_tree, 
                   feature_names = data_df.columns, 
                   class_names = ['positive', 'negative'], 
                   rounded = True, 
                   filled = True, 
                   fontsize = 14)

ConfusionMatrixDisplay.from_predictions(y_test, y_hat, cmap = 'magma')
```

Our model performed extremely well. Of a total of 52 feature vectors in our dataset, we misslabeled 1. While this model did not perform as well as the MLP, it is much easier to explain and is much more intuitive for the clients the models are developed for. Consider the medical scenario presented with this dataset. If a doctor is provided with the decision tree displayed above, they can intuitively see how the model arrives at its conclusions and is much more likely to trust and believe the model's predictions. With the MLP, although more accurate, it is difficult to explain how the model arrives at its conclusions and is very unintuitive without studying how the model operates.
