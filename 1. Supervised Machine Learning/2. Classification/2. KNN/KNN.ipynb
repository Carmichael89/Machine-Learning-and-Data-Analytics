# Phase 1 - Data Cleaning/Analysis


```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import ConfusionMatrixDisplay

from sklearn.neighbors import KNeighborsClassifier

data_df = pd.read_csv('diabetes_data_upload.csv')

data_df.head()
```

Above is a snippet of our un-modified data as a pandas dataframe. This data is not in a useful format and needs to be changed to be used in a machine learning model. We will start by encoding each feature with a 0 or 1 instead of a string to represent postive/negative, male/female, etc. We will also standardize the age value to provide a more consistent value range with the rest of the features. In this module, we will use scikit-learn's standardscalar function to cleanly and easily standardize the values. Finally, we need to append a column of ones to act as the bias of our model.

```python
data_df = data_df.replace(['Male', 'Yes', 'Positive'], 1)
data_df = data_df.replace(['Female', 'No', 'Negative'], 0)
data_df.head()
```

```python
scaler = StandardScaler()
data_df[['Age']] = scaler.fit_transform(data_df[['Age']])
data_df.head()
```

Above is the results of our cleaning. The data is now useful, the bias is now a part of our features, and the age feature is normalized. 

---
Next, we need to shuffle and split the data, train and test our model, then view the metrics. Using sci-kit learn, this process becomes incredible simple. The [scikit-learn website](https://scikit-learn.org/stable/) provides a model, a data shuffle and split method, and convenient methods to display the confusion matrix and f score. 

```python
X_train, X_test, y_train, y_test = train_test_split(data_df.iloc[:,:-1], data_df.iloc[:,-1], test_size = int(data_df.shape[0]*0.1), train_size = int(data_df.shape[0]*0.9), random_state=4)
print(X_train.shape)
print(X_test.shape)
```

# Phase 2&3: Train and Test the Model
Next, we will create a KNN model using sci-kit learn's KNeighborClassifier module and fit the data to the model. The classication module has several parameters which can be used to fine-tune the model. The parameter of the highest interest to this repository is n_neighbors, which sets the value on how many closest neighbors the model should compute. We will change and observe the impacts of this value to determine the best selection for this dataset. The other parameters are either outside of the scope of this notebook or will not make a noticible impact on the quality of our model. 




```python
for neighbors in range(1,10):
    clf = KNeighborsClassifier(n_neighbors=neighbors)
    clf.fit(X_train, y_train)
    print(f"{neighbors} Neighbors F Score: {np.round(f1_score(y_test, clf.predict(X_test)),2)}")
```

As seen above, the highest accuracy comes from calculating the nearest 1 or 3 neighbors. Calculating the nearest 1 neighbor could cause unintended sensitivity if datapoints begin to encroach on eachothers space, and higher neighbor counts can cause instability in regions where neighbors are far from the datapoint in question. For this model, we will select 3 neighbors because it provides the same accuracy and balances the potential instability of low and high neighbor calculations

```python
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train, y_train)

y_hat = clf.predict(X_test)

print("F1 score: ", np.round(f1_score(y_test, y_hat),2))

plt.figure(figsize = (50,50))

ConfusionMatrixDisplay.from_predictions(y_test, y_hat, cmap = 'magma')
```

Our model performed extremely well. Of a total of 52 feature vectors in our dataset, we misslabeled 1. While this model did not perform as well as the MLP, it is much easier to explain and is much more intuitive for the clients the models are developed for. The KNN Classifier works by calculating the K-nearest neighbors and taking a majority vote on the classification. Like the decision trees, it is easy to explain and show how the model predicts its values without overcomplicated equations.
